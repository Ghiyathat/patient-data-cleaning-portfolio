```python
import pandas as pd
import numpy as np
from dateutil import parser
```

```python
# Load file
df = pd.read_excel("Messy Patient Data.xlsx")
df
```

```python
# To clean and standardize data
def clean_date(val):
    if pd.isna(val):
        return np.nan
    try:
        if isinstance(val, (int, float)):
            return pd.to_datetime("1899-12-30") + pd.to_timedelta(val, unit="D")
        elif isinstance(val, str):
            val = val.replace("(estimated)", "").strip()
            return parser.parse(val, dayfirst=False, fuzzy=True)
        else:
            return pd.to_datetime(val, errors="coerce")
    except Exception:
        return np.nan

        # Clean all date columns
date_cols = [
    "Date_of_Admission",
    "Date_of_Diagnosis",
    "Date_of_Treatment_Start",
    "Date_of_Surgery",
    "Date_of_Discharge",
    "Follow_up_Visit"
]
for col in date_cols:
    df[col] = df[col].apply(clean_date)

    # Save cleaned data to new excel
df.to_excel("Cleaned_Patient_Data.xlsx", index=False)
    
```
